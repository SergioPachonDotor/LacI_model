```python
h=2
tmax=10
dt=0.001
t=0#tiempo
y=0#condicion inicial

tarray=[]
yarray=[]



while t<tmax:
    if y==0:
        if np.random.rand()<h*dt:
            y=1
    yarray.append(y)
    tarray.append(t)
    t+=dt
```

```python
h=2
tmax=10
dt=0.01

toccarray=[]
experimentos=10000
for exp in tqdm(range(experimentos)):
    t=0#tiempo
    y=0#condicion inicial
    while t<tmax:
        if y==0:
            if np.random.rand()<h*dt:
                toccarray.append(t)
                y=1
        t+=dt
```

```python
plt.hist(toccarray,bins=50,normed=True)
x=np.linspace(0,5,1000)
plt.plot(x,h*np.exp(-h*x),lw=2)
```

```python

kr=100
gammar=10
t=0#tiempo
r=0#condicion inicial
tarray=[]
rarray=[]
samp_time=0.01
tref=0

while t<tmax:
    if r>0:
        R=np.random.rand()
        tcrea=-(1/kr)*np.log(R)
        R=np.random.rand()
        tdeg=-(1/(gammar*r))*np.log(R)
        tau=np.min([tcrea,tdeg])
        reac=np.argmin([tcrea,tdeg])
    else:
        R=np.random.rand()
        tau=-(1/kr)*np.log(R)
        reac=0
        
    if t+tau<tref:
        if reac == 0:
            r+=1
        else:
            r+=-1
        t+=tau
    else:
        t=tref       
        tarray.append(t)
        rarray.append(r)
        tref+=samp_time    
plt.plot(tarray,rarray)
```

```python
s0=1
mu=np.log(2)
k=10*np.log(2)


t=0#tiempo
y=0#condicion inicial


tarray=[]
yarray=[]
samp_time=0.001
tref=0

while t<tmax:
    R=np.random.rand()
    tocc=(1/mu)*np.log(1-(mu*np.log(R)/(k*s0*np.exp(mu*t))))
    if t+tocc<tref:
        y+=1
        t+=tocc
    else:
        t=tref       
        tarray.append(t)
        yarray.append(y/(s0*np.exp(mu*t)))
        tref+=samp_time    
plt.plot(tarray,yarray)
```

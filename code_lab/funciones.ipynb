```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
N = 1000
x = np.linspace(-10,10, num=N)
x_2 = np.linspace(-10,10, num=N)

def f(x):
    return x ** 2

def g(x):
    return x ** 3

c = 10
y = g(-x)
y_2 = g(x)

plt.plot(x,y,'r')
plt.plot(x_2,y_2,color='grey',alpha=0.4,linewidth=5)
plt.grid()
plt.axhline(y=0, color='k',alpha=0.5)
plt.axvline(x=0, color='k',alpha=0.5)
```

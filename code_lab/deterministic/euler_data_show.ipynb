```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data_off = pd.read_csv(f'deterministic_steady_state_euler_off.csv', delimiter='|')
data_on = pd.read_csv(f'deterministic_steady_state_euler_on.csv', delimiter='|')
```

```python
time = [i/2 for i in range(120)]
plt.plot(time,data_on.y, 'g-',ms=0.8, label='Initial State On')
plt.plot(time,data_off.y, 'r--',ms=0.8, label='Initial State Off')
plt.legend(loc=7)
plt.title('Permease Deterministic\nSteady State')
plt.ylabel('Permease Concentration')
plt.xlabel('TMG (ÂµM)')
plt.xlim(0)
plt.ylim(0)
plt.grid()
```

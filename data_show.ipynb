```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
tmg = 40
cells = 1
state = 'off'
algorithm = 'gillespie'
data = pd.read_csv(f'./data/simulations/cells_{cells}_tmg_{tmg}_state_{state}_{algorithm}.csv', delimiter='|')
```

```python
plt.plot(data.Time,data.Promoter_State, '-',linewidth=0.5,ms=3, alpha=1)
plt.ylabel('Promoter State')
plt.xlabel('Time (min)')
plt.title('Promoter State')
plt.grid()
```

```python
plt.plot(data.Time,data.mRNA, 'r',linewidth=0.5,ms=0.2)
plt.ylabel('mRNA')
plt.xlabel('Time (min)')
plt.title('mRNA')
plt.grid()
```

```python

mu = '(M)'
plt.plot(data.Time,data.Permease, 'g-',linewidth=0.8,ms=0.1)
plt.ylabel(f'Permease Concentration')
plt.xlabel('Time (min)')
plt.title('Permease')
plt.grid()
plt.ylim(0)
plt.xlim(0)
```

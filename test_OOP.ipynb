```python
from Sims import Simulation as sim

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = '.tests/test_OOP_division.csv'

toy = sim(
            TMG=20, 
            cells=1, 
            file='.tests/test_OOP_division', 
            tmax=720, 
            steps=10,
            state='on'
            )
toy.simulate_adder(birth_size=1, division_time=18)
```

```python
df = pd.read_csv(file, delimiter=',')
```

```python
[plt.plot(df['Time'][df['Cell'] == i],df['Size'][df['Cell'] == i], 'grey', lw=1, alpha=0.5) for i in range(1, toy.cells + 1)];
# plt.ylim(0, 1.5)
```

```python
[plt.plot(df['Time'][df['Cell'] == i],df['mRNA'][df['Cell'] == i]/df['Size'][df['Cell'] == i], 'r', lw=0.4, alpha=0.8) for i in range(1, toy.cells + 1)];
plt.title(r'mRNA Concentration ($\frac{mRNA}{Size}$)')
plt.ylabel('mRNA')
plt.xlabel('Time (min)')
plt.grid(True, alpha=0.5)
```

```python
[plt.plot(df['Time'][df['Cell'] == i],df['Permease'][df['Cell'] == i]/df['Size'][df['Cell'] == i], 'k', lw=0.3, alpha=0.5) for i in range(1, toy.cells + 1)];
plt.title(r'Permease Concentration ($\frac{Permease}{Size}$)')
plt.ylabel('Permease')
plt.xlabel('Time (min)')
plt.grid(True, alpha=0.5)
# plt.yscale('log')
```

```python
[plt.plot(df['Time'][df['Cell'] == i],df['Permease'][df['Cell'] == i], 'k', lw=0.3, alpha=0.5) for i in range(1, toy.cells + 1)];
plt.title('Permease')
plt.ylabel('Permease')
plt.xlabel('Time (min)')
plt.grid(True, alpha=0.5)
```

```python
[plt.plot(df['Time'][df['Cell'] == i],df['mRNA'][df['Cell'] == i], 'r', lw=0.4, alpha=0.8) for i in range(1 , toy.cells + 1)];
plt.title('mRNA')
plt.ylabel('mRNA')
plt.xlabel('Time (min)')
plt.grid(True, alpha=0.5)
```

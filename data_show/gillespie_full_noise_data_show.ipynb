```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
tmg = 20
cells = 1

file_on = f'../simulation_data/gillespie_full_noise/gillespie_full_noise_on_{tmg}_cells_{cells}.csv'
file_off = f'../simulation_data/gillespie_full_noise/gillespie_full_noise_off_{tmg}_cells_{cells}.csv'

data_on = pd.read_csv(file_on, delimiter='|')
data_off = pd.read_csv(file_off, delimiter='|')
```

```python
plt.plot(data_on.Time,data_on.Permease, 'g-',ms=0.8, label='Initial State On')
#plt.plot(data_off.Time,data_off.Permease, 'r--',ms=0.8, label='Initial State Off')
plt.legend(loc=7)
plt.title(f'Permease at {tmg} M of TMG')
plt.ylabel('Permease Concentration')
plt.xlabel('Time (min)')
plt.xlim(0)
plt.ylim(0)
plt.grid()
```

```python
plt.plot(data_on.Time,data_on.mRNA, 'go',ms=0.8, label='Initial State On', linewidth=0.5)
plt.plot(data_off.Time,data_off.mRNA, 'ro',ms=0.8, label='Initial State Off',linewidth=0.5)
plt.legend(loc=1)
plt.title(f'mRNA at {tmg} M of TMG')
plt.ylabel('mRNA Concentration')
plt.xlabel('Time (min)')
plt.xlim(0)
plt.ylim(0)
plt.grid()
```

```python
plt.plot(data_on.Time,data_on.Intracellular_TMG, 'g-',ms=0.8, label='Initial State On')
plt.plot(data_off.Time,data_off.Intracellular_TMG, 'r--',ms=0.8, label='Initial State Off')
plt.legend(loc=7)
plt.title('Intracellular TMG')
plt.ylabel('Intracellular TMG Concentration')
plt.xlabel('Time (min)')
plt.xlim(0)
plt.ylim(0)
plt.grid()
```

```python
tmg = 10
cells = 1000
file = '../simulation_data/division_adder/adder_tmg_10_bs_1_divtime_18_off_1000.csv'
# df = pd.read_csv(f'../simulation_data/division_adder/adder_{tmg}_1_18.csv', delimiter='|')
df = pd.read_csv(file, delimiter='|')
df.head()
```

```python
[plt.plot(df['Time'][df['Cell'] == i], df['Size'][df['Cell'] == i], 'grey', lw=0.1, alpha=0.5) for i in range(1, cells)];
plt.grid(True, axis='y')
plt.title(f'Size')
plt.xlabel('Time (min)')
plt.ylabel('Size')
```

```python
[plt.plot(df['Time'][df['Cell'] == i], df['mRNA'][df['Cell'] == i], 'r', lw=0.1, alpha=0.5) for i in range(1, cells)];
plt.grid(True, axis='y')
plt.title(f'mRNA Levels')
plt.xlabel('Time (min)')
plt.ylabel('mRNA')
# plt.ylim(0,1)
```

```python
[plt.plot(df['Time'][df['Cell'] == i], df['Permease'][df['Cell'] == i], 'g', lw=0.1, alpha=0.5) for i in range(1, cells)];
plt.grid(True, axis='y')
plt.title(f'Permease Levels')
plt.xlabel('Time (min)')
plt.ylabel('Permease')
```
